```python
#we are using the following libraries for efficient use of data structures and data visualization
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#pandas is used for efficient use of the data frames and structured data 
df=pd.read_csv("/content/Mini Project - Marks Adding.csv") #we have added the csv file which contains all the dataset used in the program
df.head()
#we are reading the file from the headof the dataset

#we are dividing the dataset and creating dataframe which is a very efficient data structure
```

```python
df.count
#this function counts the rows and columns of the dataset, in the given data set we have 2 columns and we have displayed that below
```

```python
df['Sum']=df.groupby(['Roll Num'])['Marks'].transform('sum')
#this is a very important step, here we are defining a data frame with the same Sum and we are grouping it by using Roll Num and its corresponding marks , marks are being added using the "transform(sum)" function
print(df.drop_duplicates(subset=['Roll Num'],keep='last'))
#The duplicated data is dropped and only the first copy of the data is left
```
